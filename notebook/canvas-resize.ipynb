```python
# Resize Canvas
from PIL import Image 
import os

image_dir = "/media/data/masking/B4u/img"
part_dir = image_dir.split('/')
type_dir = part_dir[-1]
images = os.listdir(image_dir)
root_dir = '/media/data/masking/TEST/'
a = 512
b = 896

x = 400
y = int(x/a*b)

offset_x = int((1024-x)/2)
offset_y = 50

for img in images:
    save_folder = f'{root_dir}{x}/{type_dir}/'
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)
    img1 = Image.open(r"/media/data/masking/back.jpg") 
    img2 = Image.open(f'{image_dir}/{img}') 
    reimg2=img2.resize((x,y))
    
    img1.paste(reimg2, (offset_x,offset_y)) 
    
    img1.save(f'{save_folder}/{img}')
```

```python
# Resize Canvas v2
from PIL import Image 
import os

flip = True
image_dir = "/media/data/masking/89-closeup-fine/face/"
part_dir = image_dir.split('/')
type_dir = part_dir[-2]
images = os.listdir(image_dir)
root_dir = '/media/data/masking/89-closeup-fine/res/'

# x = 400

# offset_x = int((1024-x)/2)
# offset_y = 50

# size = ['250','300','350','400','450','500','550','600','650']
size = ['250']

for x in size:
  x = int(x)
  if flip:
    save_folder = f'{root_dir}{x}/{type_dir}-fliped/'
  else:
    save_folder = f'{root_dir}{x}/{type_dir}/'

  if not os.path.exists(save_folder):
      os.makedirs(save_folder)

  offset_x = int((1024-x)/2)
  offset_y = int((1024-x)/10)

  for img in images:
  
    img1 = Image.open(r"/media/data/masking/back.jpg") 
    img2 = Image.open(f'{image_dir}/{img}')
  
    a,b = img2.size
  
    y = int(x/a*b)
  
    reimg2=img2.resize((x,y))

    if flip:
      reimg2 = reimg2.transpose(Image.FLIP_LEFT_RIGHT)
    
    img1.paste(reimg2, (offset_x,offset_y),reimg2)

    # if flip:
    #   img1.transpose(Image.FLIP_LEFT_RIGHT)
    #   img1.save(f'{save_folder}/{img}')
    # else:
    img1.save(f'{save_folder}/{img}')


print('done')
```

```python
# Merge Mask

from PIL import Image 
import os

image_folder = '/home/nick/webui/ComfyUI/output/93-body/face/'
mask_folder = '/home/nick/webui/ComfyUI/output/93-body/body/'
save_folder = '/home/nick/webui/ComfyUI/output/93-body/mask/'

if not os.path.exists(save_folder):
    os.makedirs(save_folder)

image_files = os.listdir(image_folder)
mask_files = os.listdir(mask_folder)

for image_file in image_files:
  if image_file in mask_files:
    img1 = Image.open(f'{image_folder}{image_file}') 
    img2 = Image.open(f'{mask_folder}{image_file}') 

    img1.paste(img2, (0,0),img2)

    img1.save(f'{save_folder}{image_file}')

print('done')

```

```python
# Merge Mask 3

from PIL import Image 
import os

image_folder = '/home/nick/webui/ComfyUI/output/93-body/acc/'
mask_folder = '/home/nick/webui/ComfyUI/output/93-body/body/'
mask_folder2 = '/home/nick/webui/ComfyUI/output/93-body/face/'
save_folder = '/home/nick/webui/ComfyUI/output/93-body/mask/'

if not os.path.exists(save_folder):
    os.makedirs(save_folder)

image_files = os.listdir(image_folder)
mask_files = os.listdir(mask_folder)
mask_files2 = os.listdir(mask_folder2)

for image_file in image_files:
  if image_file in mask_files:
    img1 = Image.open(f'{image_folder}{image_file}') 
    img2 = Image.open(f'{mask_folder}{image_file}') 
    img3 = Image.open(f'{mask_folder2}{image_file}') 

    img1.paste(img2, (0,0),img2)
    img1.paste(img3, (0,0),img3)

    img1.save(f'{save_folder}{image_file}')

print('done')

```

```python
from PIL import Image
import numpy as np

# Create gradient image
img_width = 256
img_height = 256
gradient = np.linspace(0, 100, img_height)
gradient = np.tile(gradient, (img_width, 1))
gradient = gradient.T 
# Convert to PIL image
img = Image.fromarray(np.uint8(gradient))

# Save image
img.show()
```
